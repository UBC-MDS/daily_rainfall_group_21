5. Train ensemble machine learning model using RandomForestRegressor and evaluate with metric of your choice (e.g., RMSE) by considering Observed as the target column.

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

rmse_RF  = mean_squared_error(y_test, y_pred, squared=False)
print(f'RMSE of random forest ensemble model: {rmse_RF}')
```

```python
names = X_test.columns.to_list()

rmse = []
for name in names:
    rmse.append(mean_squared_error(y_test, X_test[name], squared=False))

results = {'models': names, 'RMSE' : rmse}
results_df = pd.DataFrame(results)
```

```python
results_df = results_df.append({'models': 'RandomForestRegressor', 'RMSE': rmse_RF}, ignore_index=True)
```

```python
results_df.sort_values("RMSE")
```
